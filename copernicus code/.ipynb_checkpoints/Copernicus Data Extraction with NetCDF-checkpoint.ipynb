Plot all daily snapshots

```python
(n_snapshots, w, h) = ds['CHL'][:,:,:].data.shape
fig, axs = plt.subplots(math.ceil(n_snapshots / 4), 4, figsize = (20, math.ceil(n_snapshots / 4) * 4))

for m, d, i in zip(ds['CHL'][:,:,:].data, ds['time'][:].data, range(n_snapshots)):
  ax = axs[i // 4]
  zsd = m
  date = datetime.date(1900, 1, 1) + datetime.timedelta(days=int(d))
  arr = (255 - np.where(zsd == -999, 0, zsd*0.03) / zsd.max() * 255).astype(np.uint8)
  img = Image.fromarray(arr,mode='L')
  img = img.resize((w, w))
  ax.set_title("%s (day %s)"%(date, i + 1))
  ax.imshow(img)
  if i > n_snapshots: 
    break
```

```python
# frames between transitions
n_frames = 3

with imageio.get_writer('uk_water_tranparency.gif', mode='I') as writer:
  for m, d, i in zip(ds['CHL'][:,:,:].data, ds['time'][:].data, range(n_snapshots)):
    if i < 18 : 
      continue
    if i > 48 :
      break 
    zsd = m
    date = datetime.date(1900, 1, 1) + datetime.timedelta(days=int(d))
    arr = (255 - np.where(zsd == -999, 0, zsd) / zsd.max() * 255).astype(np.uint8)
    img = Image.fromarray(arr,mode='L')
    img = img.resize((w, w))
    plt.title("%s (day %s)"%(date, i + 1))
    plt.imshow(img)
    filename = "snapshot_%s.png"%(i)
    plt.savefig(filename)
    image = imageio.imread(filename)
    for n in range(n_frames):
      writer.append_data(image)    
    os.remove(filename)
```
