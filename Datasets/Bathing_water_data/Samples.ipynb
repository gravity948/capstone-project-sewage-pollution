```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('samples.csv')
df
```

```python
mean_counts = df.groupby('label')[['intestinalEnterococciCount', 'escherichiaColiCount']].mean()
mean_counts
```

```python
df['discountable'].value_counts()
```

```python
df['Equal_intestinalEnterococci'] = df['intestinalEnterococciQualifier'] == '='
df['Equal_escherichiaColi'] = df['escherichiaColiQualifier'] == '='

relation_discountable = df.groupby(['Equal_intestinalEnterococci', 'Equal_escherichiaColi'])['discountable'].mean()
relation_discountable
```

```python
df['discountable'] = df['discountable'].astype(int)
mean_discountable = df.groupby('label')['discountable'].mean().reset_index()
mean_discountable = mean_discountable.sort_values('discountable', ascending=False)
mean_discountable
```
