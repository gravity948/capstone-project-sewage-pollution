```python
df = pd.read_csv('site.csv')
df.head()
```

```python
df['pollutionRiskForecasting'] = df['pollutionRiskForecasting'].map({True: 1, False: 0})
df['pollutionRiskForecasting'].mean()
```

```python
df.groupby('district')['pollutionRiskForecasting'].mean()
```

```python
df.groupby('district')['pollutionRiskForecasting'].mean().sort_values().plot(kind='bar', figsize=(20,10))

plt.xlabel("District")
plt.ylabel("pollutionRiskForecasting")

plt.show()
```
