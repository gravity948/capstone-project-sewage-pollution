```python
import pandas as pd
```

```python
riskforecasting = pd.read_csv('~data/prf.csv')
riskforecasting = riskforecasting[['EUBWID', 'predictedAt', 'warning', 'riskLevelLabel']]
riskforecasting.columns = ['site', 'time', 'warning', 'riskLevelLabel']
riskforecasting = riskforecasting[(riskforecasting['time'] >= '2022-03-01') & (riskforecasting['time'] <= '2022-10-31')]
riskforecasting['time'] = pd.to_datetime(riskforecasting['time']).dt.date
riskforecasting
```

```python
riskforecasting = riskforecasting.sort_values(by=['site', 'time', 'riskLevelLabel'], ascending=[True, True, True])
riskforecasting = riskforecasting.drop_duplicates(subset=['site', 'time'], keep='first')
riskforecasting.reset_index(drop=True, inplace=True)
riskforecasting
```

```python
riskforecasting.to_csv('~data/pollution_risk_forecasting.csv', index=False)
```
