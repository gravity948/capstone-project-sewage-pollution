```python
import pandas as pd
```

```python
riskforecasting = pd.read_csv('~data/pollution_risk_forecasting_raw.csv')
riskforecasting = riskforecasting[['label', 'predictedAt', 'warning', 'riskLevelLabel']]
riskforecasting.columns = ['site', 'time', 'warning', 'riskLevel']
riskforecasting['time'] = pd.to_datetime(riskforecasting['time']).dt.date
riskforecasting
```

```python
riskforecasting = riskforecasting.sort_values(by=['site', 'time', 'riskLevel'], ascending=[True, True, True])
riskforecasting = riskforecasting.drop_duplicates(subset=['site', 'time'], keep='first')
riskforecasting.reset_index(drop=True, inplace=True)
riskforecasting
```

```python
riskforecasting.to_csv('~data/pollution_risk_forecasting_cleaned.csv', index=False)
```
